# MD simulaion with ASE
The Atomic Simulation Environment (ASE) is a powerful Python library for setting up, manipulating, and analyzing atomistic simulations. ASE provides tools to create and visualize molecular structures, perform geometry optimizations, calculate energies and forces, and run molecular dynamics simulations. It serves as an interface to various computational chemistry codes and can be extended with custom calculators, making it a versatile tool for computational materials science and chemistry.

```python
from classical_simulation import (
    create_structure,
    attach_calculator,
    run_md_simulation,
    plot_energy,
    calculate_rdf,
    plot_rdf,
    draw_system,
    plot_3d_system,
)

# Step 1: Create the FCC structure
atoms = create_structure(structure_type="fcc", element="Cu", size=(3, 3, 3))

# Step 2: Attach a calculator
attach_calculator(atoms)

# Step 3: Visualize the structure
# Interactive viewer
draw_system(atoms)

# Static 3D plot
plot_3d_system(atoms)

# Step 4: Run molecular dynamics simulation
energies = run_md_simulation(
    atoms,
    steps=50,
    timestep_fs=5,
    energy_print_interval=10,
    temperature_K=300,
)

# Step 5: Plot energy evolution
plot_energy(energies, timestep_fs=5)

# Step 6: Calculate RDF using MDAnalysis
r, g_r = calculate_rdf(atoms, r_max=10.0, nbins=100)

# Step 7: Plot RDF
plot_rdf(r, g_r)
```
